```python
def get_file_download_url(item):
    url = item['url']
    file_result = requests.get(url)
    if file_result.status_code == 200:
        return file_result.json()['download_url']
    else:
        return ''

file_urls = []
if result['total_count']:
    items = result['items']
    file_urls = list(map(get_file_download_url, items))
```

```python
def get_file_contents(url):
    print(url)
    r = requests.get(url)
    if r.status_code == 200:
        return r.text
    else:
        return ''

file_contents = list(map(get_file_contents, file_urls))
```

```python

def find_all_addresses(file_contents):
    addresses = []
    for file in file_contents:
        for line in file.splitlines():
            search_results = re.findall("0x([a-zA-Z0-9]*)", line)
            search_results = list(filter(lambda x: len(x) >= 40, search_results))
            if search_results:
                for result in search_results:
                    addresses.append(result)
    return addresses
                
addresses = find_all_addresses(file_contents)
```

```python
def unique(seq, idfun=lambda x: x.lower()): 
   # order preserving
   if idfun is None:
       def idfun(x): return x
   seen = {}
   result = []
   for item in seq:
       marker = idfun(item)
       # in old Python versions:
       # if seen.has_key(marker)
       # but in new ones:
       if marker in seen: continue
       seen[marker] = 1
       result.append(item)
   return result
```

```python
unique_addresses = unique(addresses)
```
